```python
x = range(1,11,1)
cxMarks= [8,9,6,3,5,7,4,1,2,6]
cyMarks= [5,8,9,6,3,2,4,8,8,9]


# write a code to generate below graph
plt.plot(x,cxMarks,marker="^",color="r",ls="--")
plt.plot(x,cyMarks,marker="o",color="b",ls=":")
plt.annotate(xy=[8,1],text="Lowest SPI")
plt.legend(['cx','cy'],loc=3)
plt.xlabel("ROLL NO")
plt.ylabel("SPI")
# plt.show()
plt.savefig("mychart.jpg")
```

### 01) WAP to demonstrate the use of Pie chart.

```python
x = [10,20,30,40]
y = ['food','acomodation','travel','shopping']
c = ['r','g','y','m']
explode = [0.5,0.0,0.0,0.0]
plt.pie(x,labels=y,colors=c,explode=explode,shadow=True,autopct="%.01f%%",textprops={"fontsize":12,"color":"b"},labeldistance=1.3)
plt.title("expences")
```

### 02) WAP to to Plot List random of X, Y Coordinates in Matplotlib.

### 03) WAP to demonstrate the use of Bar chart.

```python
x = ["python","c","c++","java"]
y = [85,70,60,82]
c = ['r','g','y','m']
plt.bar(x,y,color=c,align="edge",edgecolor="b",alpha=0.5)
plt.title("popularity of languages")
plt.xlabel("languages")
plt.ylabel("ratings")
plt.show()
```

```python
x = ["python","c","c++","java"]
y = [85,70,60,82]
c = ['r','g','y','m']
plt.barh(x,y,color=c)
plt.title("popularity of languages")
plt.xlabel("languages")
plt.ylabel("ratings")
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x = ["python","c","c++","java"]
y = [85,70,60,82]
c = ['r','g','y','m']
plt.bar(x,y,color=c,edgecolor="b")
def addLabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
plt.title("popularity of languages")
plt.xlabel("languages")
plt.ylabel("ratings")
addLabels(x,y)
plt.show()
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
c = ["r","b","g","r","b"]
width = [1,2,3,4,5]
plt.scatter(x,y,color=c,marker="*",linewidths=width)
plt.xlabel("x")
plt.ylabel("y")
plt.title("star")
plt.show()
```
